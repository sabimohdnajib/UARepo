```python
import requests
import pandas as pd
import geopandas as gpd

# Describing the url
url_bikes = "https://api.glasgow.gov.uk/mobility/v1/get_rentals?startDate=2022-05-01&endDate=2023-05-01"
# Making the query to the web server, using the Get method from the requests library 
response = requests.get(url_bikes)
response
```

```python
#Get the response from the web server, need to translate that into a format that can be manipulated, like JSON.
data = response.json()
data #Cell out was cleared as file was large
```

```python
#Using the data label to get all attributes included. 
rental_data = data['data']
rental_data
#Cell output cleared
```

```python
rental_pd = pd.DataFrame(rental_data)
rental_pd.head() #Creating a dataframe from the data on rental bikes
```

```python
# Checking for NaN in the coordinates column
nan_in_column_Lat = rental_pd['startPlaceLat'].isna().any()
nan_in_column_Long = rental_pd['startPlaceLong'].isna().any()

print(nan_in_column_Lat,nan_in_column_Lat) #There are missing values for both columns on longitude and latitude
```

```python
clean_rental_pd = rental_pd.dropna(subset=['startPlaceLat', 'startPlaceLong', 'endPlaceLat','endPlaceLong'])
clean_rental_pd.info() #Cleaning the dataframe so there are no missing values
```

```python
#Creating a GeoDataFrame from a DataFrame using coordinates
gdf_bikes_end = gpd.GeoDataFrame(clean_rental_pd, geometry=gpd.points_from_xy(clean_rental_pd['endPlaceLong'], clean_rental_pd['endPlaceLat']))
```

```python
#Looking at the bikes end data
gdf_bikes_end.info()
```

```python
#Converting data in columns to different types and formats
gdf_bikes_end.endPlaceId = gdf_bikes_end.endPlaceId.astype(int)
gdf_bikes_end.endPlaceName = gdf_bikes_end.endPlaceName.astype(str)
gdf_bikes_end['endDate'] = pd.to_datetime(gdf_bikes_end['endDate'], format='%Y-%m-%dT%H:%M:%SZ')

gdf_bikes_end.head()
```

```python
#Only keeping relevant columns
keep_cols_end = [
    "endDate",
    "endPlaceId",
    "endPlaceName",
    "durationSeconds",
    "isInvalid",
    "price",
    "isEbike",
    "endPlaceLat",
    "endPlaceLong",
    "geometry",
] #only keeping specific columns necessary
gdf_bikes_end = gdf_bikes_end[keep_cols_end]
gdf_bikes_end.head()
```

```python
#Adjusting the long and lat values for gdf_bikes_end
from sklearn.cluster import KMeans
num_clusters = 4

kmeans_collection_end = KMeans(n_clusters=num_clusters, random_state=42)
gdf_bikes_end['kmeans_cluster_end'] = kmeans_collection_end.fit_predict(gdf_bikes_end[['endPlaceLong', 'endPlaceLat']])
```

```python
#Generating the map for gdf_bikes_end
import leafmap
m_end = leafmap.Map(center=(55.860166, -4.257505),
                zoom=12,
                draw_control=False,
                measure_control=False,
                fullscreen_control=False,
                attribution_control=True,
                   
               )

m_end.add_basemap("CartoDB.Positron")
m_end.add_data(
    gdf_bikes_end,
    column='kmeans_cluster_end',
    legend_title='Clusters',
    cmap='Set1',
    k=4,
)

#Ploting the map
m_end
```
